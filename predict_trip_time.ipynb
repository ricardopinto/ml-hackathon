```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.callbacks import EarlyStopping
import pandas as pd
import numpy as np
import datetime
```

```python
# prepare dataset by merging weather data into trip data

df_trips = pd.read_csv('trip.csv', index_col='trip_id', parse_dates=['starttime', 'stoptime'])
df_trips.head()

df_weather = pd.read_csv('weather.csv', index_col='Date', parse_dates=['Date'])
df_weather.head()
```

```python
weather_cols = df_weather.columns.drop(['Events', 'Max_Gust_Speed_MPH']).values
weather_cols
```

```python
df = df_trips.copy()
df['Date'] = df['starttime'].apply(lambda x: datetime.date(year=x.year, month=x.month, day=x.day).isoformat())
```

```python
df[weather_cols] = df['Date'].apply(lambda d: df_weather.loc[d,weather_cols])
```

```python
df.head()
```

```python
df_weather.loc['2014-10-13':'2014-10-15',weather_cols]
```

```python
df.columns
```

```python
features = df.drop([
    'Date',
    'gender',
    'tripduration',
    'starttime',
    'stoptime',
    'bikeid',
    'from_station_name',
    'to_station_name',
    'from_station_id',
    'to_station_id',
    'usertype',
    'gender',
    'birthyear'
], axis=1).columns.values

target = 'tripduration'
validation_split = 0.2
dataset_size = df.shape[0]
features
```

```python
df_train  = df.copy()
df_train = df_train[features]
df_train = df_train.fillna(df_train.mean())
```

```python
x_train = df_train.as_matrix()
y_train = df[target].values
```

```python
n_cols = len(features)  # or x_train.shape[1]
input_shape = (n_cols,)

# Specify the model
model = Sequential()
model.add(Dense(100, activation='relu', input_shape = input_shape))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='relu'))


# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Define early_stopping_monitor
early_stopping_monitor = EarlyStopping(patience=5)

# Fit the model
model.fit(
    x_train,
    y_train,
    epochs=10,
    validation_split=validation_split
    #callbacks=[early_stopping_monitor]
)
```

```python
pred = model.predict(np.array([df_train.iloc[0]])) # 985
pred[:,0][0]
```
